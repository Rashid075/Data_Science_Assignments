# Hypothesis Testing

### 1. State the Hypotheses statement



*   The theoretical weekly operating cost model is W=1000+5X
*   Given, the number of units produced pper week has MuX=600


*   Expected weekly cost is MuW=1000+5(600)= 4000
*   Null Hypothesis(H0): The mean weekly operating cost is equal to the theoretical cost. i.e H0: MuW = 4000

*   Alternative Hypothesis(H1): The mean weekly operating cost is greater than the theoretical cost. i.e H1: MuW > 4000








### 2. Calculate the Test Statistic

```python
import pandas as pd
import numpy as np
```

```python
# Given Values
x_bar = 3050 #sample mean weekly cost
mu = 1000 + 5*600 #theoretical mean weekly cost
sigma = 5 *25 #standard deviation of weekly cost
n=25 #sample size
```

```python
#calculating the standard error
std_error = sigma/np.sqrt(n)
std_error
```

```python
#calculating the z-statistic
z_stats = x_bar - mu/std_error
z_stats
```

```python
# creating dataframes
results_df = pd.DataFrame({
    "Sample Mean (x̄)": [x_bar],
    "Theoretical Mean (μ)": [mu],
    "Population Std Dev (σ)": [sigma],
    "Sample Size (n)": [n],
    "Standard Error": [std_error],
    "Z Test Statistic": [z_stats]
})
results_df
```

### 3. Determining the Probability

```python
from scipy import stats
```

```python
# Given Values
z = -38
alpha = 0.05
```

```python
#Right-tailed p-value
p_value = 1 - stats.norm.cdf(z)
p_value
```

### 4. Making a Decision



*   alpha = 0.05
*   z = -38 (test statistic)

*   z(0.05) = 1.645
*   Decision: Reject H0 if z>1.645

*   since -38 < 1.645
*   So we failed to reject the null hypothesis







### 5. Conclusion

At the 5% significance level, there is no evidence to support the restaurant owners’ claim that the weekly operating costs are higher than predicted by the cost model. In fact, the observed sample mean cost is significantly lower than the theoretical mean.
