# Exercise 1: Prime Numbers

```python
#to check whether a given number is prime number or not.
num=int(input("Enter a number:"))
if num<=1:
  print("Not a Prime Number")
else:
  is_prime=True
  for i in range(2, int(num**0.5)+1):
    if num%i==0:
      is_prime=False
      break
  if is_prime:
    print(f"{num} is a Prime Number ")
  else:
    print(f"{num} is not a Prime Number")
```

# Exercise 2: Product of Random Numbers

```python
import random
ran_1= random.randint(1,10)
ran_2= random.randint(1,10)
user_inp=int(input(f"Enter the product of {ran_1} and {ran_2}:"))
res=ran_1 * ran_2
if user_inp == res:
  print("Correct Answer!!")
else:
  print("Wrong Answer!!")
  print(f"The correct answer is {res}")
```

# Exercise 3: Squares of Even/Odd Numbers

```python
# to check the squares of even numbers
print("The Squares of even numbers from 100 to 200/n:")
for i in range(100, 201):
  if i%2==0:
    square=i**2
    print(f"The Square of the {i} is {square}")
```

# Exercise 4: Hospital Billing System

```python
# to calculate the final bill

#hospital charges
room_charges_per_day= 2000
doctor_fee=1500
lab_test_charges=300

#patient usage
no_of_days=4
no_of_tests=3
medicine_charges=2400

room_chareges= no_of_days * room_charges_per_day
test_charges= no_of_tests * lab_test_charges

total_bill= room_chareges + doctor_fee + test_charges + medicine_charges
print(f"The total bill before discount = {total_bill}")
if total_bill > 10000:
  discount = total_bill * 0.10
else:
  discount = 0

final_bill = total_bill - int(discount)
print(f"The Final bill after discount = {final_bill}")
```
