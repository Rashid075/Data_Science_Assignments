# Basic Statistics

### Import the required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Reading the CSV file

```python
sales = pd.read_csv('sales_data_with_discounts.csv')
sales.head()
```

### Identifying the Numerical and Categorical Columns

```python
#Numerical Columns
num_cols = sales.select_dtypes(include=['int64', 'float64']).columns
num_cols
```

```python
#categorical Columns
cat_cols = sales.select_dtypes(include=['object']).columns
cat_cols
```

### Now calculating the Mean, Median, Mode, Standard Deviation to the Numerical columns

```python
#calculatig the mean
# mean represents the average value of the data
sales[num_cols].mean()
```

```python
#calculating the median
#median indicates the middle value and is useful when data contains outliers
sales[num_cols].median()
```

```python
#calculating the mode
#mode displays the most frequently occuring values
sales[num_cols].mode()
```

```python
#calculating the Standard Deviation
# standard deviation measures the spread of the data
sales[num_cols].std()
```

-> If the mean is higher than the median, the data is positively skewed.

-> A large standard deviation indicates high variability in values.

-> The mode helps identify commonly occurring values.

-> Comparing these measures provides insights into the overall distribution.

### Data Visualization

#### To visualize the distribution and relationship of numerical and categorical variables in the dataset.

#### Histogram

```python
#plotting histogram
sales[num_cols].hist(bins=20, figsize=(12,8))
plt.suptitle('Histogram of Numerical Columns')
plt.show()
```

-> Histograms are used to show how numerical data is distributed by grouping values into bins and displaying their frequencies. Separate histograms are plotted for each numerical columns.

-> The histogram shows that most numerical columns are right-skewed.

-> Total Sales Value, Net Sales Value, Discount Amount, and Avg Price have many values clustered at the lower end with a few very high values.

-> This indicates that most transactions are small, but there are some large sales, which act as outliers.

-> Overall, the data is not normally distributed and contains skewness and extreme values.

#### Box Plot

```python
# plotting box plot
#sales[num_cols].plot(kind='box', figsize=(12,8)) #one way
sales[num_cols].boxplot(figsize=(12, 8)) #another way
plt.suptitle('Box Plot of Numerical Columns')
plt.show()
```

-> The box plots clearly highlight the presence of several outliers in:

  Total Sales Value

Net Sales Value

Discount Amount

Avg Price

-> The outliers appear as individual points beyond the whiskers, indicating extreme values.

-> This shows high variability in sales and discount-related values.

-> Overall, the box plots confirm that the dataset contains unusual high values and is not evenly distributed.

#### Bar Chart Analysis for Ctegorical Columns

```python
# bar chart analysis
for col in cat_cols:
  plt.figure()
  sales[col].value_counts().plot(kind='bar')
  plt.title(f'Bar Chart of {col}')
  plt.xlabel(col)
  plt.ylabel('Count')
  plt.show()
```

-> Bar charts are used to show how frequently each category appears in the dataset.

-> From the above bar chart:

   => Categories with taller bars occur more frequently.

   => Categories with shorter bars are less common.
   
   => An uneven bar height shows that some categories dominate the dataset.

#### Conclusion

-> From the descriptive analytics and visualizations, we can clearly understand the given dataset.

-> The statistical measures showed that sales, discounts, and prices have high variation and are positively skewed, meaning most values are small but a few are very large.

-> Histograms and box plots helped identify outliers, especially in sales value and discount amount, which represent unusually large transactions.

-> Bar charts of categorical data showed that some categories occur much more frequently than others, indicating imbalanced distributions.
Overall, these analyses give a clear overview of the data and help prepare it for further analysis or modeling.
